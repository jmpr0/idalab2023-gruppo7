## TAKE HOME

- Increasing the training set size reaches a performance plateau at $\approx 24$ classes, thus we selected 24 + 8 + 8 as training, validation, test split. N.B. the number of epochs (i.e. $100-300$) does not impact the outcome of analysis
- Classification performance increase with the increasing of the number of shots
- Using a powerful features extractor (w.r.t. silhouette score), i.e. thre Hybrid CNN+RNN by Lopez, negatively affects classification performance ($10%$ drop in F1)
